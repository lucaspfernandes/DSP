```python
import numpy as np
import matplotlib.pyplot as plt
```

## Impulso

```python
lim = 10
N = int(10)
t = np.linspace(0, lim, N+1)     
```

```python
d = lim/N

def impulse(t, n, delta = d):
    ans = np.zeros(t.shape)
    try:
        ans[(n / delta).astype(int)] = 1.0
    except: 
        ans[int(n / delta)] = 1.0
    return ans
```

```python
fig = plt.figure(figsize=(20,5))
plt.stem(t, impulse(t, np.array([3,4,5])))
plt.show()
```

## Degrau

```python
lim = 10
N = int(10)
t = np.linspace(0, lim, N+1)
```

```python
d = lim/N

def step(t, n, delta = d):
    ans = np.zeros(t.shape)
    s = int( n / delta )
    ans[s:] = np.ones((len(t)-s))
    return ans
```

```python
fig = plt.figure(figsize=(20,5))
plt.stem(t, step(t, 6))
plt.show()
```
